```python
%%R
install.packages("agricolae")
library(agricolae)

# Your data for DMRT, LSD, and Tukey's test
data <- data.frame(
  treatment = rep(c("T1", "T2", "T3", "T4", "T5", "T6", "T7", "T8"), each=3),
  value = c(12, 15, 8, 20, 12, 17, 14, 21, 17, 3, 2, 6, 28, 29, 36, 43, 37, 39, 0, 0, 0, 96, 91, 87)
)

# Perform ANOVA
model <- aov(value ~ treatment, data = data)
summary(model)

# Perform DMRT
dmrt <- duncan.test(model, "treatment", console=TRUE)
print(dmrt)

# Perform LSD test
lsd <- LSD.test(model, "treatment", console=TRUE)
print(lsd)

# Perform Tukey's test
tukey <- HSD.test(model, "treatment", console=TRUE)
print(tukey)

# Save results to a variable and print
results <- capture.output({
  print("ANOVA Result")
  print(summary(model))
  print("DMRT Result")
  print(dmrt$statistics)
  print(dmrt$groups)
  print("LSD Result")
  print(lsd$statistics)
  print(lsd$groups)
  print("Tukey's Test Result")
  print(tukey$statistics)
  print(tukey$groups)
})

print(results)

# Save the results to a text file
sink("results.txt")
cat(results, sep = "\n")
sink()
```
